```python
#incase of duplicate payment ids
payment_id_duplicates = df[df.duplicated('payment_id', keep=False)]
df = df.drop_duplicates(subset=['payment_id'], keep='first')
```

```python
#checkin negative payments
negative_amounts = df[df['amount'] < 0]
```

```python
#where there are negative i have changed the transaction to be refund or credit this is solely on my assumptions that there are no error s and the busin3ess is falling so many are taking refunds
df['transaction_type'] = np.where(df['amount'] < 0, 'refund/credit', 'payment')
```

```python
#a good measure would be to add a subscription length
df['subscription_length'] = (df['valid_until'] - df['payment_date']).dt.days
```

#Overlaps

```python
user_counts = df['user_id'].value_counts()
overlapping_users = user_counts[user_counts > 1].index.tolist()
print(len(overlapping_users))
```

```python
#so there are 1489 overlaps
#since these overlaps are so many therefore i would want to keep only the latest subscription for a user
df = df.sort_values('payment_date', ascending=False).drop_duplicates('user_id')

```
