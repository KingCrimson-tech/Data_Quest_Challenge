```python
#by the above data i think the valid range of ratings is 1 to 5 and rest are ridiculous
min = 1
max = 5
valid = (df['rating'] >= min) & (df['rating'] <= max)
```

```python
df_cleaned = df[valid].copy()
```

```python
df_cleaned
```

```python
#so around a 1000 were invalid
```

```python
#conflicting ratings are those which are like duplicates and i would the same as in watch_history i would keep the latest one and drop others
df_cleaned['review_date'] = pd.to_datetime(df_cleaned['review_date'])
```

```python
df_most_recent = df_cleaned.sort_values('review_date').drop_duplicates(
    subset=['user_id', 'movie_id'], 
    keep='last'
)
```
